```python
import openai
import base64
from IPython.display import display
from PIL import Image


#chave de autenticação
openai.api_key = "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"

response = openai.Image.create(
    prompt="crie a imagem de um iate cor de rosa no céu",
    size="512x512",
    n=1,
    response_format="b64_json"
)
#recebe a resposta
imagem = response["data"][0]["b64_json"]

#transforma em imagem
imagem = base64.b64decode(imagem)

with open("computador.png", "wb") as f:
    f.write(imagem)

display(Image.open("computador.png"))

```
